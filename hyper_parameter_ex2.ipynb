```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
dir(df)
```

```python
data_frame = pd.DataFrame(df.data,columns=df.feature_names)
data_frame.head()
```

```python
data_frame['target'] = df.target
data_frame['target'] = data_frame['target'].apply(lambda x:df.target_names[x])
data_frame.head()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.data,df.target,test_size=0.3)
```

```python
svm = SVC(kernel ='linear',C =10,gamma ='auto')
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(SVC(kernel='linear',C=10,gamma='auto'),df.data,df.target)
```

```python
from sklearn.model_selection import GridSearchCV
gd = GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['linear','rbf']
},cv = 5,return_train_score=False)
gd.fit(df.data,df.target)
gd.cv_results_
```

```python
data = pd.DataFrame(gd.cv_results_)
data.head()
```

```python
data[['param_C','param_kernel','params']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['linear','rbf']
},cv =5, return_train_score=False,n_iter=2)
rs.fit(df.data,df.target)
rs.cv_results_
```

```python
data = pd.DataFrame(rs.cv_results_)
data
```

```python
model_params = {
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['linear','rbf']
        }
    },
     'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators': [1, 10, 20],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4],
            'max_features': ['auto', 'sqrt', 'log2']
            
        }
    },
     'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,10,20],
            
        }
    },
     'gaussian_nb':{
        'model':GaussianNB(),
        'params':{
            'priors': [None], 
            'var_smoothing': [1e-9, 1e-6, 1e-3]
            
        }
    },
    'multinomial_nb':{
        'model':MultinomialNB(),
        'params':{
           'alpha': [1.0, 0.5, 0.1],  
            'fit_prior': [True, False]
            
        }
    },
     'decision_tree':{
        'model':DecisionTreeClassifier(),
        'params':{
           'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4],
            'max_features': ['auto', 'sqrt', 'log2']
            
        }
    }
}
    
       
```

```python
scores = []
for model_name,params in model_params.items():
    model = params['model']
    params_grid = params['params']
    
    gd = GridSearchCV(model,params_grid,cv =5,return_train_score=False)
    gd.fit(df.data,df.target)
    
    scores.append({
        'model':model_name,
        'best_score':gd.best_score_,
        'best_params':gd.best_params_
    })
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
